```python
import os
import io
import json
import requests
import pandas as pd
import numpy as np
import torch
import torchvision
import transformers
from typing import Dict, List, Any, Optional

class AdShieldMultiPlatformAnalyzer:
    def __init__(self, config_path='config.json'):
        """
        Initialize AdShield Multi-Platform Scam Detection System

        Args:
            config_path (str): Path to configuration file with API credentials
        """
        # Load configuration
        try:
            with open(config_path, 'r') as config_file:
                self.config = json.load(config_file)
        except FileNotFoundError:
            self.config = {}

        # Initialize AI Models
        self.text_model = transformers.BertForSequenceClassification.from_pretrained('bert-base-uncased')
        self.text_tokenizer = transformers.BertTokenizer.from_pretrained('bert-base-uncased')

        # Scam Categories
        self.scam_categories = {
            0: 'Phishing',
            1: 'Investment Fraud',
            2: 'Fake Products',
            3: 'Identity Theft',
            4: 'Romance Scam',
            5: 'Technical Support Scam'
        }

        # Media Processing Setup
        self.media_processors = {
            'text': self.process_text
        }

    def collect_social_media_data(self, platforms: List[str], keywords: List[str]) -> pd.DataFrame:
        """
        Collect simulated data from multiple social media platforms

        Args:
            platforms (List[str]): Social media platforms to simulate
            keywords (List[str]): Scam-related keywords to search

        Returns:
            pd.DataFrame: Collected advertisement data
        """
        collected_data = []

        # Simulated data collection for each platform
        sample_data = [
            {
                'platform': 'instagram',
                'content_type': 'text',
                'text': 'Get Rich Quick with our Crypto Investment!',
                'timestamp': pd.Timestamp.now()
            },
            {
                'platform': 'facebook',
                'content_type': 'text',
                'text': 'Guaranteed Returns - Investment Opportunity',
                'timestamp': pd.Timestamp.now()
            },
            {
                'platform': 'twitter',
                'content_type': 'text',
                'text': 'Amazing offer - Limited Time Technical Support Services',
                'timestamp': pd.Timestamp.now()
            }
        ]

        # Filter data based on platforms and keywords
        for item in sample_data:
            if item['platform'] in platforms and \
               any(keyword.lower() in item['text'].lower() for keyword in keywords):
                collected_data.append(item)

        return pd.DataFrame(collected_data)

    def process_media(self, media_item: Dict) -> Dict[str, Any]:
        """
        Process and analyze different media types

        Args:
            media_item (Dict): Media item details

        Returns:
            Dict with scam analysis results
        """
        processor = self.media_processors.get(media_item.get('content_type', 'text'))
        if processor:
            return processor(media_item)
        return {'error': 'Unsupported media type'}

    def process_text(self, text_data: Dict) -> Dict[str, Any]:
        """
        Analyze text for potential scam content

        Args:
            text_data (Dict): Text advertisement data

        Returns:
            Dict with text analysis results
        """
        text = text_data.get('text', '')

        # Tokenize and process text
        inputs = self.text_tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)

        with torch.no_grad():
            outputs = self.text_model(**inputs)
            probabilities = torch.softmax(outputs.logits, dim=1)

        predicted_class = torch.argmax(probabilities).item()

        return {
            'text': text,
            'scam_category': self.scam_categories[predicted_class],
            'confidence_score': probabilities[0][predicted_class].item()
        }

def main():
    # Initialize AdShield
    adshield = AdShieldMultiPlatformAnalyzer()

    # Define search keywords
    scam_keywords = [
        'investment', 'crypto', 'guaranteed returns',
        'free money', 'limited offer'
    ]

    # Collect data from multiple platforms
    collected_data = adshield.collect_social_media_data(
        platforms=['instagram', 'facebook', 'twitter'],
        keywords=scam_keywords
    )

    # Analyze collected data
    scam_analysis_results = []
    for _, media_item in collected_data.iterrows():
        analysis_result = adshield.process_media(media_item.to_dict())
        scam_analysis_results.append(analysis_result)

    # Display Results
    for result in scam_analysis_results:
        print("\nScam Detection Result:")
        print(json.dumps(result, indent=2))

# Run the main function
if __name__ == "__main__":
    main()
```
